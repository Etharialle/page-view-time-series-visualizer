```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
df = pd.read_csv('fcc-forum-pageviews.csv', index_col='date', parse_dates=True)
```

```python
df['month'] = df.index.month
df['year'] = df.index.year

df.head()
```

```python
df = df.groupby(['year', 'month']).mean().reset_index()
df.head()
```

```python
df.head(15)
df.dtypes
```

```python
month_legend = {
    1: 'January',
    2: 'February',
    3: 'March',
    4: 'April',
    5: 'May',
    6: 'June',
    7: 'July',
    8: 'August',
    9: 'September',
    10: 'October',
    11: 'November',
    12: 'December'}
df.replace({'month': month_legend}, inplace=True)
df.head(15)
```

```python
sns.barplot(data=df, x='year', y='value', hue='month', palette='bright')
plt.show()
```
